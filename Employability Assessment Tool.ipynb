# Final Project 2 - Employability Assessment Tool

The dataset **candidates.xlsx** contains the test result of students who gave an online test for job.

**The confidentiality of this dataset has been maintained by changing names, email address and userid of all the students.**

The test consisted of several questions divided in 5 different sections. This dataset has total score as well as section wise scores of each student.

You can download the data from : "https://tinyurl.com/sitip19-python-final-project"

### Dataset description

- Alloted Set - It contains number of set from 1 to 5.
- Total Score - It contains the total marks scored by the student.

There are 5 different sections and 3 feature columns (given below) for every sections.

- Section **n** - Total questions attempted
- Wrong Answers **n** - Incorrect attempts
- Marks **n** - Correct attempts

(n stands for section number)

```python
f=0
g=0
h=0
i=0
j=0
s=df['Marks 1'].max()
for name,values in df.groupby('Name'):
    if all(s==values['Marks 1']):
        f+=1
        print("Name of topper in section 1 is {0}".format(name))
print("there are {0} toppers".format(f))
p=df['Marks 2'].max()
for name,values in df.groupby('Name'):
    if all(p==values['Marks 2']):
        g+=1
        print("Name of topper in section 2 is {0}".format(name))
print("there are {0} toppers".format(g))
q=df['Marks 3'].max()
for name,values in df.groupby('Name'):
    if all(q==values['Marks 3']):
        h+=1
        print("Name of topper in section 3 is {0}".format(name))
print("there are {0} toppers".format(h))
r=df['Marks 4'].max()
for name,values in df.groupby('Name'):
    if all(r==values['Marks 4']):
        i+=1
        print("Name of topper in section 4 is {0}".format(name))
print("there are {0} toppers".format(i))
t=df['Marks 5'].max()
for name,values in df.groupby('Name'):
    if all(t==values['Marks 5']):
        j+=1
        print("Name of topper in section 5 is {0}".format(name))
print("there are {0} toppers".format(j))
```
